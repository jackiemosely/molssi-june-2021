# Accessing Data

```python
# To slice a pandas dataframe using row and column numbers
# dataframe_name.iloc[row_slice,column_slice]

periodic_data.iloc[35]
```

```python
periodic_data.iloc[35:45, :3]
```

```python
# Use the iloc function to:
# 1. Select the 5th row
# 2. Select row 30 to the end and the first 3 columns
# 3. Select row 50 to the end and every other row and columns 2 through 4
```

```python
periodic_data.iloc[4]
```

```python
periodic_data.iloc[30:, :3]
```

```python
periodic_data.iloc[50::2,2:5]
```

```python
# To get a column of information
# dataframe_name["column_name"]

# List the column names
periodic_data.columns
```

```python
periodic_data["AtomicMass"]
```

```python
periodic_data[["AtomicMass", "Symbol"]]
```

```python
# Third way to get information from datafrrames - loc function
# loc function lets us use column and row names

periodic_data.loc[35, "Symbol"]
```
